#**How many cars have horsepower less than 100?**

```python
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce').astype('Int64')#converted horsepower from object datatype to Int64
```

```python
df.info()
```

```python
df[df['horsepower']<100][['make']]
```

```python
df[df['horsepower']<100][['make']].value_counts()#this code shows the breakup of count of cars by the make
```

```python
df[df['horsepower']<100][['make']].count()#There are a total of 109 cars
```
